## Demo of the pipeline of maize extraction
### renaming the files based on QR code
### croping the each ear from the raw images


```python
import os 

def find_images_folders(root_folder):
    image_folders = []
    for dirpath, dirnames, filenames in os.walk(root_folder):
        for dirname in dirnames:
                image_folders.append(os.path.join(dirpath, dirname)) 
    return image_folders
```

```python
PATH_HYBRID_FOLDER = "Data/hybrid_maize"
```

```python
image_folders = find_images_folders(PATH_HYBRID_FOLDER)
```

```python
from utils.qr_reader import rename_files_based_on_qr_code
```

```python
import concurrent.futures
def parallel_process_folders(folders):
    with concurrent.futures.ThreadPoolExecutor() as executor:
        executor.map(rename_files_based_on_qr_code, folders)
```

```python
parallel_process_folders(image_folders)
```

```python
CROP_PATH_HYBRID_FOLDER = "Data/hybrid_maize/Croped"
```

```python
import os
def get_image_paths(root_folder):
    # List of image extensions
    image_extensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp', '.tiff']
    
    # List to store image paths
    image_paths = []

    # Walk through the directory
    for subdir, _, files in os.walk(root_folder):
        for file in files:
            # Check if the file is an image
            if any(file.lower().endswith(ext) for ext in image_extensions):
                # Add the file path to the list
                image_paths.append(os.path.join(subdir, file))
    
    return image_paths

# Example usage
root_folder = 'Data/hybrid_maize/Raw'  # Replace with the actual path to folder A
image_paths = get_image_paths(root_folder)
```

```python
from utils.crop import Crop
```

```python
def process_image(image_path, crop_path):
    crop = Crop(image_path, crop_path)
    crop.crop_image()
```

```python
from concurrent.futures import ThreadPoolExecutor
with ThreadPoolExecutor() as executor:
    futures = [executor.submit(process_image, image_path, CROP_PATH_HYBRID_FOLDER) for image_path in image_paths]
    
    # Optionally wait for all futures to complete
    for future in futures:
        future.result()
```
